```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_blobs
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering

# Function to plot clustering results
def plot_clusters(X, labels, title):
    unique_labels = np.unique(labels)
    for label in unique_labels:
        plt.scatter(
            X[labels == label, 0], X[labels == label, 1], 
            label=f'Cluster {label + 1}', s=50, cmap='viridis'
        )
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

```

```python
# Dataset 1: make_moons 
X_moons, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
# Apply DBSCAN
dbscan_moons = DBSCAN(eps=0.2, min_samples=5).fit(X_moons)
# Apply k-Means
kmeans_moons = KMeans(n_clusters=2, random_state=42, n_init=10).fit(X_moons)
# Apply Hierarchical Clustering
hier_moons = AgglomerativeClustering(n_clusters=2).fit(X_moons)

# Plot clustering results
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1)
plot_clusters(X_moons, dbscan_moons.labels_, 'DBSCAN on Moons')

plt.subplot(2, 3, 2)
plot_clusters(X_moons, kmeans_moons.labels_, 'k-Means on Moons')

plt.subplot(2, 3, 3)
plot_clusters(X_moons, hier_moons.labels_, 'Hierarchical Clustering on Moons')
```

```python
# Dataset 2: make_blobs
X_blobs, _ = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)

# Apply DBSCAN
dbscan_blobs = DBSCAN(eps=1.0, min_samples=5).fit(X_blobs)
# Apply k-Means
kmeans_blobs = KMeans(n_clusters=3, random_state=42, n_init=10).fit(X_blobs)
# Apply Hierarchical Clustering
hier_blobs = AgglomerativeClustering(n_clusters=3).fit(X_blobs)

# Plot clustering results
plt.subplot(2, 3, 4)
plot_clusters(X_blobs, dbscan_blobs.labels_, 'DBSCAN on Blobs')

plt.subplot(2, 3, 5)
plot_clusters(X_blobs, kmeans_blobs.labels_, 'k-Means on Blobs')

plt.subplot(2, 3, 6)
plot_clusters(X_blobs, hier_blobs.labels_, 'Hierarchical Clustering on Blobs')

plt.tight_layout()
plt.show()
```
